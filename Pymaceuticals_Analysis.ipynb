# Pymaceuticals Analysis

## Takeaways

- Among the four drugs (Capomulin, Infubinol, Ketapril, and Placebo), Capomulin was the only effective drug that decreased the size of tumor over the span of 45 days. 

- Both Capomlin and Infubinol were effective controling metastatic spread that after 45 days of treatment, the number of metastatic sites in mice treated by the two drugs were significantly lower than mice treated with Placebo which was the control group (standard error bar not overlapped). 

- Mice treated with Capomulin had the highest survival rate after 45 days of treatment (~21%). Mice treated with Infubinol had the lowest survival rate, lower than mice treated with Placebo dug. 

## Data Analysis

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug = "data/mouse_drug_data.csv"
clinical_trial = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse = pd.read_csv(mouse_drug)
df_clinical = pd.read_csv(clinical_trial)
```

```python
#display mouse data
df_mouse.head()
```

```python
#check data info
df_mouse.info()
#check if each value is unique
df_mouse.nunique()
```

```python
#one mouse received two treatments. Need to find the ID of the mouse and drop value
mouse_drop = df_mouse.groupby(['Mouse ID']).count()
mouse_ID = mouse_drop[mouse_drop['Drug']==2]
mouse_ID
```

```python
df_mouse_cleaned = df_mouse[df_mouse['Mouse ID'] !='g989']
```

```python
#display clinical data
df_clinical.head()
```

```python
#check data info
df_clinical.info()
df_clinical.nunique()
```

```python
#outerjoin two dataset
df_final = df_mouse_cleaned.merge(df_clinical,"outer",on='Mouse ID',sort='Mouse ID')
df_final.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume_mean = df_final.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()
# Convert to DataFrame
tumor_volume_mean_df = pd.DataFrame(tumor_volume_mean).reset_index()
# Preview DataFrame
tumor_volume_mean_df.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_ste = df_final.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
tumor_volume_ste_df = pd.DataFrame(tumor_volume_ste).reset_index()
# Preview DataFrame
tumor_volume_ste_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_volume_mean_df_pivot = tumor_volume_mean_df.pivot_table(index = 'Timepoint',
                                                              columns='Drug',
                                                              values='Tumor Volume (mm3)')
# Preview that Reformatting worked
tumor_volume_mean_df_pivot.head(10)
```

```python
#Make a list of drug to be researched (Capomulin, Infubinol, Ketapril, and Placebo)
#Make a list of markers and colors to be usedd in graph for the four drugs respectively
drug_list = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
marker_list = ['o','^','s','D']
color_list = ['red','blue','green','black']

#using for loop to create the four lines

fig, ax = plt.subplots()
for i in range(len(drug_list)):
    ax.errorbar(tumor_volume_mean_df_pivot.index,tumor_volume_mean_df_pivot[drug_list[i]],
                yerr = tumor_volume_ste_df[tumor_volume_ste_df['Drug']==drug_list[i]]['Tumor Volume (mm3)'], 
                marker=marker_list[i],mfc=color_list[i],mec=color_list[i],linestyle='--',color=color_list[i])

ax.legend()
ax.set_title('Tumor Response to Treatment')
ax.set_xlabel('Time (Days)')
ax.set_ylabel('Tumor Volume (mm3)')
ax.grid()

fig.savefig('line1.png',dpi=100)

plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_mean = df_final.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()
# Convert to DataFrame
met_mean_df = pd.DataFrame(met_mean).reset_index()
# Preview DataFrame
met_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_error = df_final.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()
# Convert to DataFrame
met_error_df = pd.DataFrame(met_error).reset_index()
# Preview DataFrame
met_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_mean_df_pivot = met_mean_df.pivot_table(index = 'Timepoint',columns='Drug',values='Metastatic Sites')
# Preview that Reformatting worked
met_mean_df_pivot.head()
```

```python
#using for loop to create the lines

fig, ax = plt.subplots()
for i in range(len(drug_list)):
    ax.errorbar(met_mean_df_pivot.index,met_mean_df_pivot[drug_list[i]],
                yerr = met_error_df[tumor_volume_ste_df['Drug']==drug_list[i]]['Metastatic Sites'],
                marker=marker_list[i],mfc=color_list[i],mec=color_list[i],
                linestyle='--',color=color_list[i])

ax.legend()
ax.set_title('Metastatic Spread During Treatment')
ax.set_xlabel('Treatment Duration (Days)')
ax.set_ylabel('Met. Sites')
ax.grid()

fig.savefig('line2.png',dpi=100)

plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = df_final.groupby(['Drug','Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count).reset_index()
# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_df_pivot = mice_count_df.pivot_table(index = 'Timepoint',columns = 'Drug',values='Mouse ID')
# Preview the Data Frame
mice_count_df_pivot.head()
```

```python
#plotting

fig,ax = plt.subplots()

for i in range (len(drug_list)):
    ax.plot(mice_count_df_pivot.index,mice_count_df_pivot[drug_list[i]],
            marker=marker_list[i],mfc=color_list[i],mec=color_list[i],
            linestyle='--',color=color_list[i])

ax.legend(loc='lower left')
ax.set_title('Survival During Treatment')
ax.set_xlabel('Time (Days)')
ax.set_ylabel('Survival Rate (%)')
ax.grid()

fig.savefig('line3.png',dpi=100)

plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
change = (tumor_volume_mean_df_pivot.iloc[9,:]-tumor_volume_mean_df_pivot.iloc[0,:])/tumor_volume_mean_df_pivot.iloc[0,:]

# Display the data to confirm
change
```

```python
#create x axis and y axis value

x_axis = [i for i in range (len(drug_list))]
y_axis = [change[drug_list[i]] for i in range (len(drug_list))]

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()

ax.bar(x_axis,y_axis,color=['red'if i>=0 else 'green' for i in y_axis],tick_label=drug_list)

labels = ['{:.2%}'.format(i) for i in y_axis]

height = [0.03 if y>0 else -0.03 for y in y_axis]

for i in range (len(labels)):
    ax.text(x_axis[i],height[i],labels[i],
            horizontalalignment='center',verticalalignment='center',
            color='white')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_ylabel('% Tumor Volume Change')
ax.grid()

fig.savefig('bar.png')

plt.show()
```
